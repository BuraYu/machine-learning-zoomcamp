```python
df = pd.read_csv('data.csv')
```

Sometimes, it's necessary to clean a little bit of data. In this case we are removing all the whitespace and add _ + make everything lower case

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.head()
```

Shows which datatype each column has.

```python
df.dtypes
```

```python
df.dtypes == 'object'
```

To only select everthing that is Object

```python
df.dtypes[df.dtypes == 'object']
```

since we need access to the index only: 

```python
strings = list(df.dtypes[df.dtypes == 'object'].index)
```

Since it's a list we can loop over them

```python
for col in strings:
    df[col] = df[col].str.lower().str.replace(' ', '_')
```
